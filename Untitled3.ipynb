```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "adult.csv"

df = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "uciml/adult-census-income",
  file_path

)
```

```python
# Print columns
print(list(df))
# Replace NaN
import numpy as np
df = df.replace('?', np.nan)
```

```python
import matplotlib.pyplot as plt
df.isna().sum().plot(kind='bar', figsize=(10, 6), color='coral')
plt.title('Figure 1.')
plt.xlabel('Unknown values per column.')
plt.tight_layout()
```
